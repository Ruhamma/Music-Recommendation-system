Clustering the Data

Means =[] => The number of clusters (k) for each iteration in the loop
intertias=[] => The sum of th squares of the clusters-> to know how tightly the data points are clusteres around the centroids.

-> optimise_k_means(data, max_k) => data(the data set we want to consider) && max_k(The maximum number of clusters to consider)

-> create an empty array for storing means(number of clusters) & intertias

-> Inertia=> A measure of how well the data points are clustered. 
          => It is the sum of squared distances of samples to their closest cluster center.                  
          => Lower inertia means that the data points are closer to their respective centroids, indicating more compact and well-defined clusters.
          
-> for k in range(1, max_k + 1): => iterate over a range of cluster numbers from 1 to max_k

-> kmeans = KMeans(n_clusters=k, n_init=10) => initialize the kmeans algorithm with k clusters
                                            => fit the kmeans model to the data
                                            
                     
->plt.figure(figsize=(10, 5))=> size of the plot

->plt.plot(means, inertias, 'o-'): Plot the number of clusters (k) on the x-axis and the corresponding inertia on the y-axis. The 'o-' specifies a line plot with circle markers at each data point.


```python
def optimise_k_means(data, max_k):
    means = []
    inertias = []

    for k in range(1, max_k + 1):
        kmeans = KMeans(n_clusters=k, n_init=10)
        kmeans.fit(data)
        means.append(k)
        inertias.append(kmeans.inertia_)

    plt.figure(figsize=(10, 5))
    plt.plot(means, inertias, 'o-')
    plt.title('Elbow Method For Optimal k')
    plt.xlabel('Number of Clusters (k)')
    plt.ylabel('Inertia')
    plt.show()


optimise_k_means(numeric_df, 10)
```

```python
def find_track_index(track_name, df):
    
    if 'Track' not in df.columns:
        raise ValueError("DataFrame does not have a 'Track' column")
    
    
    try:
        track_index = df.loc[df['Track'] == track_name].index[0]
        return track_index
    except IndexError:
        return None



def find_song_recommendation(track_name, df, num_recommendations=5):
    
    track_index = find_track_index(track_name, df)
    
    
    if track_index is None:
        print(f"Track '{track_name}' not found in the DataFrame")
        return None
    
    
    cluster = df.loc[track_index, 'Cluster']
    
   
    filtered_df = df[df['Cluster'] == cluster]
    filtered_df
    
   
    if filtered_df.empty:
        print("No similar tracks found in the same cluster")
        return None
    
   
    for i in range(num_recommendations):
        recommendation = filtered_df.sample()
        track = recommendation.iloc[0]['Track']
        artist = recommendation.iloc[0]['Artist']
        print(f"{track} by {artist}")



find_song_recommendation('blue', df)
```
